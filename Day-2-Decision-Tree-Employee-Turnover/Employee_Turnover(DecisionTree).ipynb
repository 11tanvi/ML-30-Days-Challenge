**Import Libraries**

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
```

**Load dataset**

```python
df = pd.read_csv("employee_turnover.csv")
```

```python
df.head()
df.columns
df.isnull().sum()
df.info()
df.nunique()
```

```python
x = df.drop("Employee_Turnover",axis = 1)
x.head()
```

```python
y = df["Employee_Turnover"]
y.head()
```

**EDA**

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm")
```

**Train Test Split**

```python
#train test split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state = 42)
```

```python
x.isnull().sum()
```

**DecisionTreeClassifier**

```python
# Decision Tree Model
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
y_pred = model.predict(x_test)
print("accuracy",accuracy_score(y_test,y_pred))
```

**Tree Plot**

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

plt.figure(figsize=(12,8))
plot_tree(model, filled=True)
```

**Accuracy**

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,confusion_matrix
print("accuracy",accuracy_score(y_test, y_pred)*100,"%")
print("precision",precision_score(y_test,y_pred)*100, "%")
print("recall",recall_score(y_test, y_pred)*100,"%")
print("confusion_matrix",confusion_matrix(y_test,y_pred)*100, "%")
```

**##Conclusion

In this project, Decision Tree Classifier was applied to predict employee turnover.

The model achieved:

* Accuracy: 84.44%
* Precision: 82.67%
* Recall: 84.0%
